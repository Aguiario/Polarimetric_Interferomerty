```python
import numpy as np
import matplotlib.pyplot as plt
import pyPI as pi
import cv2
```

```python
sample, sample_image, X, Y = pi.initialize_environment()
# Definir posiciones fijas para las lneas horizontales (en pxeles)
fixed_positions = [100, 250, 400]  # Posiciones en el eje y

# Definir matrices especficas para cada lnea en sample
line_matrices = [
    pi.jones_matrix(np.pi, np.pi/2),        # Lnea 1
    pi.jones_matrix(np.pi/3, np.pi/3),      # Lnea 2
    pi.jones_matrix(np.pi/8, np.pi/4)       # Lnea 3
]

sample = pi.add_lines(fixed_positions, line_matrices, sample_image, sample, grosor=50)

Eis = np.array([[1], [1j]])
Er = np.array([[1], [1j]])

info = pi.sample_intensity(Er, Eis, sample, X, Y)
```

```python
# Inicializar el entorno
sample, sample_image, X, Y = pi.initialize_environment()

# Determinar el centro de la imagen para colocar las posiciones de manera adecuada
fixed_positions = [(100, 400), (500, 100), (800, 250)]  # (x, y) en pxeles

# Definir matrices especficas para cada lnea en sample
line_matrices = [
    pi.jones_matrix(np.pi, np.pi/2),        # Lnea 1
    pi.jones_matrix(np.pi/3, np.pi/3),      # Lnea 2
    pi.jones_matrix(np.pi/8, np.pi/4)       # Lnea 3
]

# Agregar crculos en las posiciones indicadas
sample = pi.add_circles(fixed_positions, line_matrices, sample_image, sample)

Eis = np.array([[1], [1j]])
Er = np.array([[1], [1j]])

info = pi.sample_intensity(Er, Eis, sample, X, Y)
```
